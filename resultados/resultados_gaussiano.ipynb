# Recolección de tiempos de ejecución con diferentes tamaños de matrices Gaussiano

Importes necesarios

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Python

Recolectamos los resultados para el algoritmo de Python sin paralelismo

```python
#matrices de tamaños: 10, 100, 200, 300, 400, 500, 1000, 5000
tamaños = [10, 100, 200, 300, 400, 500, 1000, 5000]

resultados_python = [[0.000830, 0.000376, 0.000411, 0.000662, 0.000414],
                     [0.025509, 0.027487, 0.027911, 0.028645, 0.033091],
                     [0.115810, 0.102652, 0.105408, 0.106665, 0.110727],
                     [0.253792, 0.274481, 0.268371, 0.305001, 0.271213],
                     [0.500348, 0.535398, 0.471051, 0.467502, 0.542928],
                     [0.800366, 0.831532, 0.833589, 0.818477, 0.882411],
                     [4.877934, 4.929803, 4.995684, 4.955038, 4.903449],
                     [98.053778, 100.816497, 91.236864, 102.508435, 94.256196]]

resultados_python_paralelo = [[7.360188, 7.594391, 7.542439, 6.944589, 6.983203],
                              [7.282104, 6.876118, 6.796128, 6.835389, 7.117969],
                              [7.322597, 7.194208, 7.041000, 7.122958, 7.197001],
                              [7.004513, 6.996436, 6.885945, 7.097282, 7.519617],
                              [7.303614, 7.191631, 7.026115, 7.017148, 7.028524],
                              [7.033885, 6.945095, 6.931743, 6.881469, 7.237039],
                              [8.891140, 8.715844, 9.034407, 10.082465, 8.965964],
                              [61.780132, 62.367023, 64.184870, 64.821789, 65.156338]]
```

Calculamos el promedio y la desviación estándar

```python
tamaños = np.array(tamaños)

promedios = [np.mean(r) for r in resultados_python]
desviaciones = [np.std(r) for r in resultados_python]
promedios_paralelo = [np.mean(r) for r in resultados_python_paralelo]
desviaciones_paralelo = [np.std(r) for r in resultados_python_paralelo]

promedios = np.array(promedios)
desviaciones = np.array(desviaciones)
promedios_paralelo = np.array(promedios_paralelo)
desviaciones_paralelo = np.array(desviaciones_paralelo)
```

Graficamos la media de tiempos de inversión no paralelizada con respecto a las dimensiones de la matriz.

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios - desviaciones, promedios + desviaciones,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Graficamos la media de tiempos de inversión paralelizada con respecto a las dimensiones de la matriz.

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios_paralelo, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios_paralelo - desviaciones_paralelo, promedios_paralelo + desviaciones_paralelo,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
